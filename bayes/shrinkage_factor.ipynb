# Shrinkage factorの形状

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def lasso_sf(x, param=0.1):
    return x**(-2) * np.exp(-1 / (param * x))

params = [0.1, 1, 10]
x = np.linspace(1e-3, 1, 100)
for p in params:
    y = lasso_sf(x, p)
    y = y / np.max(y)
    plt.plot(x, y, label=str(p))
plt.title("lasso")
plt.legend(loc="lower right")
plt.savefig("pdf_lasso.png")
plt.show(9)
```

```python
def stu_t_sf(x, a, b, param):
    return x ** (a - 1) * (1 - x) ** (- a - 1) * np.exp(- param * b * x / (1 - x))

a = b = 0.5
params = [0.1, 1, 10]
x = np.linspace(1e-3, 1-1e-5, 100)
for p in params:
    y = stu_t_sf(x, a, b, p)
    y = y / np.max(y)
    plt.plot(x, y, label=str(p))
plt.title("student_t")
plt.legend(loc="lower right")
plt.savefig("pdf_student_t.png")
plt.show(9)
```

```python
def horseshoe_sf(x, param):
    return x ** (-0.5) * (1 - x) ** (-0.5) / (param * x + 1 - x)

params = [0.1, 1, 10]
x = np.linspace(1e-3, 1-1e-5, 100)
for p in params:
    y = horseshoe_sf(x, p)
    y = y / np.max(y)
    plt.plot(x, y, label=str(p))
plt.title("horseshoe")
plt.legend(loc="lower right")
plt.savefig("pdf_horseshoe.png")
plt.show()
```
