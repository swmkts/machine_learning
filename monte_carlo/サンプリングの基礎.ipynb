## 逆変換法
$$
    1. Generate \quad u: u \sim U(0, 1)
$$
$$
    2. Return \quad x: x = F^{-1}(u)
$$

```python
def generate_exp_dist(x, theta):
    return - (1 / theta) * np.log(1 - x)
x = np.linspace(-1, 5)
theta = 1.2
n_sample = 100000
y = theta * np.exp(- theta * x)
u = np.random.uniform(0.0, 1.0, n_sample)
x_trans  = - (1 / theta) * np.log(1 - u) 
plt.hist(x_trans, normed=True, bins=500, label='sample')# 必ず正規化とbinの数を調整する
plt.plot(x, y, label='theoritical')
plt.xlim(-1, 5)
plt.ylim(0, 2)
plt.legend()
plt.show()
```

## 棄却法(regection method)
$$
    1. y \sim q(y)
$$
$$
    2. u \sim U(0, 1)
$$
$$
    3. if \quad u \leq \frac{\pi(y)}{Cq(y)} \quad then \quad accept \quad y
$$

### 切断正規分布からサンプリングする
$$
    \pi (x) = \sqrt{\frac{2}{\pi}} e^{-x^2/2}
$$
### 棄却法では$1/C$が採択確率になる。$C$を小さくするような$\pi(x)$を選ぶことが重要
### ベイズ統計でも比の計算で正規化定数が不要になることが重要である
#### $l(x) \leq \pi(x)$なる$l(x)$が見つかれば抉り(えぐり)出し法が使える

```python
x = np.linspace(0, 4)
y = np.sqrt(2 / np.pi) * np.exp(- x ** 2 / 2)
beta = 1.0
C = 1 / 0.76
# iter_num = 10000
# y_accepted = []
# for i in range(0, iter_num):
#     q_y = np.random.exponential(beta, 1)
#     pi_y = np.sqrt(2 / np.pi) * np.exp(- q_y ** 2 / 2)
#     u = np.random.uniform(0, 1, 1)
#     if u <=  pi_y / (C * q_y):
#         y_accepted.append(q_y)
q_y = np.random.exponential(beta, n_sample)
u = np.random.uniform(0, 1, n_sample)
pi_q = np.sqrt(2 / np.pi) * np.exp(q_y - q_y ** 2 / 2)
y_accepted = q_y[u <=  0.76 * pi_q]
print(len(y_accepted))
plt.hist(y_accepted, bins=80, normed=True)
plt.plot(x, y)
plt.show()
```

## 適応的棄却法(adaptive rejection method)
* $\pi(x)$が複雑な時には$q(x)$が見つかりにくい
* $\pi(x)$を区分的に近似する

## 混合法(mixture method)
* 混合分布$\pi$からサンプリングする(離散)
$$
    i \sim (w_1, w_2, \cdots, w_r)
$$
$$
    x \sim \pi_i(x)
$$
* 混合分布$\pi$からサンプリングする(連続)
$$
   y \sim w(y)
$$
$$
    x \sim p(x | y)
$$
例えば、t分布$\int_0^\infty p(x|y)w(y)dy \quad x|y \sim N(0, 1/y), \quad y \sim Ga(\nu/2, \nu/2)$は混合法によりサンプリング可能な分布である

* よって、ベイズ予測分布は事後分布からパラメータをサンプリングできれば簡単にサンプリングできる

## 一様乱数比法(ratio of uniforms method)
1. $$u_1, u_2 \sim U(0, 1)$$
2. $$u = au_1, v = b_- + (b_+ - b_-)u_2$$
3. if \quad $$(u, v) \in C_p$$ then \quad return   $$x = \frac{v}{u}$$  else reject \frac{v}{u}

### 切断正規分布からサンプリングする
$$
    \pi (x) = \sqrt{\frac{2}{\pi}} e^{-x^2/2}
$$

```python
c = np.linspace(-4, 4)
c_y = c * np.sqrt(np.exp(-c**2/2))
plt.plot(c, c_y)
plt.show()
```

```python
c = np.linspace(0, 4)
c_y = np.sqrt(2 / np.pi) * np.exp(- c ** 2 / 2)
u1 = np.random.uniform(0, 1, n_sample)
u2 = np.random.uniform(0, 1, n_sample)
a = 1
b_minus = 0
b_plus = np.sqrt(2 / np.e)
u = a * u1
v = b_minus + (b_plus - b_minus) * u2
x = v / u
x_accepted = x[(0 <= u) & (u <= np.sqrt(np.exp(-(v / u) ** 2 / 2)))]
plt.plot(c, c_y)
plt.hist(x_accepted, normed=True, bins=80)
plt.plot()
```
